```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
```

```python
fig = plt.figure() #an empty figure with no axes
fig, ax= plt.subplots() #a fig with single axes
fig, ax1= plt.subplots(2,3) #fig having axes in 2 rows and 3 column
```

```python
x= np.linspace(0, 2, 100) #random data
fig, ax = plt.subplots(figsize=(5, 2.7), constrained_layout=True)

ax.plot(x, x, label='linear') #plotting data on axes
ax.plot(x, x**2, label='quadratic') #plotting data on axes
ax.plot(x, x**3, label='cubic') #plotting data on axes
ax.set_xlabel("x label") #adding title to x-axis
ax.set_ylabel("y label") #adding title to y-axis
ax.set_title("Graph Plot")
ax.legend()

```
