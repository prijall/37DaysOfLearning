```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def plotter(ax, data1, data2, param_dict):
    """
    A helper function to make graph
    """

    out= ax.plot(data1, data2, **param_dict) # **param_dict is used to unpack thr dictionary 'param_dict' into the arg of function
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100) #random dataset
fig, (ax1, ax2)= plt.subplots(1,2, figsize=(5, 2.7))
plotter(ax1, data1, data2, {'marker': 'x'})
plotter(ax2, data3, data4, {'marker': 'o'})
```
